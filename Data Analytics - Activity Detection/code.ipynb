```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import os

import warnings
warnings.filterwarnings('ignore')
```

# Cycling dataset

```python
acc = pd.read_csv('Accelerometer.csv')
gyro = pd.read_csv('Gyroscope.csv')
```

```python
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.scatter(acc['x'], acc['y'], alpha=0.5)
plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1, 3, 2)
plt.scatter(acc['x'], acc['z'], alpha=0.5)
plt.title('X vs Z')
plt.xlabel('X')
plt.ylabel('Z')

plt.subplot(1, 3, 3)
plt.scatter(acc['y'], acc['z'], alpha=0.5)
plt.title('Y vs Z')
plt.xlabel('Y')
plt.ylabel('Z')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.scatter(gyro['x'], gyro['y'], alpha=0.5)
plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1, 3, 2)
plt.scatter(gyro['x'], gyro['z'], alpha=0.5)
plt.title('X vs Z')
plt.xlabel('X')
plt.ylabel('Z')

plt.subplot(1, 3, 3)
plt.scatter(gyro['y'], gyro['z'], alpha=0.5)
plt.title('Y vs Z')
plt.xlabel('Y')
plt.ylabel('Z')

plt.tight_layout()
plt.show()
```

# Walking Dataset

```python
acc2 = pd.read_csv('acc.csv')
gyro2 = pd.read_csv('gyro.csv')
```

```python
acc2['time']=pd.to_datetime(acc2['time'], format='mixed')
```

```python
acc2[['x','y','z']].describe()
```

```python
plt.figure(figsize=(15,10))

plt.subplot(3,1,1)
plt.plot(acc2['time'],acc2['x'], color='red')
plt.title('X Movement')

plt.subplot(3,1,2)
plt.plot(acc2['time'],acc2['y'], color='green')
plt.title('Y Movement')

plt.subplot(3,1,3)
plt.plot(acc2['time'],acc2['z'], color='yellow')
plt.title('Z Movement')
```

```python
plt.figure(figsize=(15,10))

plt.subplot(3,1,1)
plt.plot(acc2['time'],acc2['x'], color='red', alpha=0.5)

plt.subplot(3,1,1)
plt.plot(acc2['time'],acc2['y'], color='green', alpha=0.5)

plt.subplot(3,1,1)
plt.plot(acc2['time'],acc2['z'], color='yellow', alpha = 0.5)

plt.title('X Y Z Movement')
```

```python
corr_max = acc2[['x','y','z']].corr()
sns.heatmap(corr_max, annot=True)
```

```python
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.scatter(acc2['x'], acc2['y'], alpha=0.5)
plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1, 3, 2)
plt.scatter(acc2['x'], acc2['z'], alpha=0.5)
plt.title('X vs Z')
plt.xlabel('X')
plt.ylabel('Z')

plt.subplot(1, 3, 3)
plt.scatter(acc2['y'], acc2['z'], alpha=0.5)
plt.title('Y vs Z')
plt.xlabel('Y')
plt.ylabel('Z')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.subplot(3,1,1)
plt.plot(gyro2['time'],gyro2['x'], color='red')
plt.title('X Movement')

plt.subplot(3,1,2)
plt.plot(gyro2['time'],gyro2['y'], color='green')
plt.title('Y Movement')

plt.subplot(3,1,3)
plt.plot(gyro2['time'],gyro2['z'], color='yellow')
plt.title('Z Movement')
```

```python
plt.figure(figsize=(15,10))

plt.subplot(3,1,1)
plt.plot(gyro2['time'],gyro2['x'], color='red', alpha=1)

plt.subplot(3,1,1)
plt.plot(gyro2['time'],gyro2['y'], color='green', alpha=0.5)
plt.title('X vs Y Movement')

plt.subplot(3,1,2)
plt.plot(gyro2['time'],gyro2['x'], color='red', alpha=1)

plt.subplot(3,1,2)
plt.plot(gyro2['time'],gyro2['z'], color='green', alpha=0.5)
plt.title('X vs Z Movement')

plt.subplot(3,1,3)
plt.plot(gyro2['time'],gyro2['z'], color='red', alpha=1)

plt.subplot(3,1,3)
plt.plot(gyro2['time'],gyro2['y'], color='green', alpha=0.5)
plt.title('Z vs Y Movement')
```
